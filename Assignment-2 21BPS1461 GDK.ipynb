###  Visualizations
### 1) Univariate Analysis

```python
plt.figure(figsize=(8, 6))
sns.distplot(df['Area of the house(excluding basement)'])
```

```python
sns.displot(df.Longitude)
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['id'])
plt.title('ID')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Date'])
plt.title('Date')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.pie((df['waterfront present']).value_counts(),[0,0.2],labels = ['NO','YES'],autopct ='%1.1f%%',shadow = False,colors = ['red','yellow'])
plt.title('Waterfront present')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.pie((df['number of views']).value_counts(),[0,0.1,0,0.1,0],labels = [4, 0, 2, 1, 3],autopct ='%1.1f%%',shadow = False,colors = ['purple','yellow','orange','green'],)
plt.title('Number of views')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x =(df['number of views']).value_counts().index,y =(df['number of views']).value_counts() )
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x =(df['number of floors']).value_counts().index,y =(df['number of floors']).value_counts() )
```
